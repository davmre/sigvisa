```python
tm_types = {"tt_residual":"constant_laplacian",
            "peak_offset":"param_linear_mb",
            "amp_transfer":"param_sin1",
            "coda_decay":"param_linear_distmb",
            "peak_decay":"param_linear_distmb"}

tvals = dict()
tvals["freq_2.0_3.0"] = dict()
tvals["freq_2.0_3.0"][5334939] = {'arrival_time': 1240246030.92,
         'peak_offset': 1.786317, 
         'coda_height': 0.593107, 
         'peak_decay': -3.38506, 
         'coda_decay': -3.149505, 
         'amp_transfer': 2.26198}

tvals["freq_2.0_3.0"][5335822] = {'arrival_time': 1240241955.57,
         'peak_offset': 2.803422, 
         'coda_height': 0.490531, 
         'peak_decay': -4.321054, 
         'coda_decay': -2.682453, 
         'amp_transfer': 2.532104}

tvals["freq_0.8_4.5"]=dict()
tvals["freq_0.8_4.5"][5334939] = {'arrival_time': 1240246031.32,
         'peak_offset': 1.359683, 
         'coda_height': 2.382346, 
         'peak_decay': -3.853422, 
         'coda_decay': -3.232372, 
         'amp_transfer': 4.072172}

tvals["freq_0.8_4.5"][5335822] = {'arrival_time': 1240241957.11,
         'peak_offset': 0.990592, 
         'coda_height': 2.038895, 
         'peak_decay': -4.035579, 
         'coda_decay': -3.232372, 
         'amp_transfer': 4.093359}

def get_graph(wave_evid, ev_evid, band="freq_2.0_3.0", wiggle_runid=29):
    sg = SigvisaGraph(template_shape = "lin_polyexp", template_model_type = tm_types,
                          wiggle_family = "db4_2.0_3_30", wiggle_model_type = "gp_lld",
                          dummy_fallback = True, nm_type = "ar",
                          runids=(26,wiggle_runid), phases=["P",], gpmodel_build_trees=False,
                          absorb_n_phases=False, uatemplate_rate=1e-20)
    print sg.wiggle_family

    w = load_event_station_chan(wave_evid, "MKAR", "auto", evtype="leb", cursor=None, pre_s = 10, post_s=200)
    w_filtered = w.filter("%s;env;hz_5.0" % band)
    wn = sg.add_wave(w_filtered)

    ev = get_event(evid=ev_evid)
    ev.depth=0.0
    evnodes = sg.add_event(ev)
    
    sg.set_template(1, wn.sta, "P", wn.band, wn.chan, tvals[band][wave_evid])

    return sg, wn

def visualize(wn, variance=False):
    wn._parent_values()
    N=400
    plt.figure(figsize=(10,5))
    v = wn.get_value()
    plot(v.data[:N], color="black", lw=1)
    #for i in range(5):
        #wn.parent_sample()
    #    plot(wn.get_value().data)
    mobs=wn.tssm.mean_obs(400)
    vobs=wn.tssm.obs_var(400)
    plot(mobs, c="green", lw=2, label="predicted")
    if variance:
        plot(mobs+2*np.sqrt(vobs), c="red", label="2 stddev")
        plot(mobs-2*np.sqrt(vobs), c="red")
    print wn.log_p()
    legend()
```
