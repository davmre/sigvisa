```python
def sample_ar(alpha, n):
    p = len(alpha)
    x = np.zeros((n+p,))
    x[:p] = np.random.randn(p)
    for i in range(n):
        x[p+i] = np.dot(alpha, x[i:p+i]) + np.random.randn()
    return x[p:]

alpha = np.array([-0.1, 0.2, 0.7])
xs = np.zeros((1000, 10))
for i in range(1000):
    xs[i,:] = sample_ar(alpha, 20)[5:15]
```

```python
cov = np.dot(xs.T, xs)
prec = np.linalg.inv(cov)


imshow(cov, interpolation='none')
figure()
imshow(prec, interpolation='none')
```

```python
def expand_ar(alpha, eps):
    n = len(eps)
    p = len(alpha)
    xs = np.zeros((n,))
    for  t in range(p):
        xs[t] = eps[t]
    
    for t in range(p, n):
        for j in range(1, p+1):
            xs[t] += xs[t-j] * alpha[j-1]
        xs[t] += eps[t]
    return xs

def ar_linear(alpha, N):
    poly = np.concatenate([(1,), -np.array(alpha)])
    roots = np.roots(poly[::-1])
    invRoots = 1.0/roots
    
    ps = np.array([np.real(np.sum(invRoots**d)) for d in range(N+1)])
    
    zs = np.zeros((N,))
    zs[0] = 1
    for n in range(1, N):
        zs[n] = 1.0/n * np.sum([ps[k] * zs[n-k] for k in range(1,n+1)])
    return zs

def expand_linear(zs, eps):
    n = len(eps)
    k = len(zs)
    xs = np.zeros((n,))
    
    for t in range(k, n):
        for j in range(1, k+1):
            xs[t] += eps[t-j] * zs[j-1]
    return xs

```

```python
def ar_autocovs(zs, n):
    def autocov(k):
        return np.sum([zs[i] * zs[i+k] for i in range(len(zs)-k)])
    
    ac = np.zeros((n,))
    for k in range(n):
        ac[k] = autocov(k)
    return ac

def ar_cov(alpha, n, z_order=100):
    zs = ar_linear(alpha, z_order)
    ac = ar_autocovs(zs, n)
    
    C = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            C[i,j] = ac[np.abs(i-j)]
    return C
    

```

```python
alpha = (.4,.2,.1,.1)
C = ar_cov(alpha, 10)
imshow(C, interpolation='none')
#print C[0,:]

prec = np.linalg.inv(C)
prec = np.where(np.abs(prec) > 1e-10, prec, np.zeros(prec.shape))
print prec
figure()
imshow(prec, interpolation='none')
```
