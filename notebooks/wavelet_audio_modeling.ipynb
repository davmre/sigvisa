```python
from sigvisa.models.wiggles.wavelets import construct_implicit_basis, construct_basis_simple
from sigvisa.models.statespace.compact_support import CompactSupportSSM, ImplicitCompactSupportSSM
```

```python
# test implicit basis code

N = 128
start_times, end_times, identities, prototypes = construct_implicit_basis(N, "db4", "zpd")
basis = construct_basis_simple(N, "db4", "zpd")

for (st, et, ident, b) in zip(start_times, end_times, identities, basis):
    if st >= 0 and et <= N:
        assert((b[st:et] == prototypes[ident]).all())
    if st >= 0:
        assert((b[:st] == 0).all())
    if et <= N:
        assert((b[et:] == 0).all())
    
cssm = CompactSupportSSM(basis)
z = cssm.prior_sample(128)
ll = cssm.run_filter(z)
icssm = ImplicitCompactSupportSSM(start_times, end_times, identities, prototypes)
ll2 = icssm.run_filter(z)
print ll
print ll2
assert(np.abs(ll-ll2) < 1e-8)
```
