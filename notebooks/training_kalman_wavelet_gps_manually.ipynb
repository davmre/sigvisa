```python
import scipy.stats
import pywt
from sigvisa.models.statespace.ar import ARSSM
from sigvisa.models.statespace.compact_support import CompactSupportSSM
from sigvisa.models.statespace.transient import TransientCombinedSSM
from sigvisa.models.noise.armodel.model import ARModel, ErrorModel
```

```python
N = 256
cc = pywt.wavedec(np.zeros(N,), 'db4', 'zpd')
nb = np.sum([len(l) for l in cc])
bases = np.empty((nb, N))

k = 0
for i,l in enumerate(cc):
    for j in range(len(l)):
        cc[i][j] = 1.0
        bases[k+j, :] = pywt.waverec(cc, 'db4', 'zpd')
        cc[i][j] = 0.0
    k += len(l)

small_bases = bases[:74, :]
```

```python
def multiply_mvn(a, A, b, B, P=None):
    Ainv = np.linalg.inv(A)
    Binv = np.linalg.inv(B)
    if P is None:
        C = np.linalg.inv(Ainv+Binv)
        c = np.dot(C, np.dot(Ainv, a) + np.dot(Binv, b))
    else:
        C = np.linalg.inv(Ainv+np.dot(P.T, np.dot(Binv, P)))
        c = np.dot(C, np.dot(Ainv, a) + np.dot(P.T, np.dot(Binv, b)))
    return c, C

params = np.array([0.6,0.1 ])
evar = 0.5
arssm = ARSSM(params, evar)
ar_model = ARModel(params, ErrorModel(0, np.sqrt(evar)))
```

```python
# first, for each train station, compute the initial ADF message.
# Note all of these messages will be diagonal, so combining them will yield a block-diagonal GP.
local_prior_K = (w_var+noise_var) * np.eye(74)
local_prior_Kinv = np.linalg.inv(local_prior_K)
KCinv = np.linalg.inv(local_prior_K+C) # (K+C)^-1
diag_KCinv_inv = np.linalg.inv(np.diag(np.diag( KCinv))) # diag( (K+C)^-1 )^-1
unstableB = diag_KCinv_inv - local_prior_K


P = np.linalg.inv(local_prior_Kinv + Cinv) # true posterior cov (K^-1 + C^-1)^-1
Pdiag = np.diag(np.diag(P)) # target posterior cov: what ADF should get us to


# more stable way to compute B?
# diag((K^-1+C^-1)^-1) = (K^-1 + B^-1)^-1
# B = (diag((K^-1+C^-1)^-1)^-1 - K^-1 )^-1
#B =  np.linalg.inv(np.linalg.inv(Pdiag) - local_prior_Kinv)

bs = []
#derived_bs = []
for j in range(len(train_stations)):
    c = base_cs[j]
    Cc = cs[j]
    p = np.dot(P, c) # true posterior mean: what adf should get to
    b, B = multiply_mvn(p, Pdiag, zero74, -local_prior_K)
    
    #Cc = cs[j]
    #b = np.dot(diag_KCinv_inv, np.dot(KCinv, Cc))
    
    bs.append(b)
    #derived_bs.append(derived_b)
    
# to predict a coef, we have
# alpha = (K+B)^-1 b
# where B and b aggregate the diagonal terms from training messages. 
# 
# where does this come from? did I just invent it?
# we should have a posterior cov P=(K^-1 + sum B^-1)^-1
# and a posterior mean p = P(sum_j B^-1 b)
# to predict from a posterior, we have
# mean = k^T K^-1 m
#      = k^T K^-1 P(sum B^-1 b)
#      where K^-1P = K^-1 (K^-1 + sum_j B^-1)^-1 = (K + inv(sum_j B^-1))^-1  inv(sum_j B^-1)
#     so the mean is
#       k^T (K + inv(sum_j B^-1))^-1 inv(sum_j B^-1) (sum_j B^-1 b)
#  here we are predicting a single param at a new location, so b_j, B_j is the message regarding this param from the j'th station. 
#  each b_j is a one-hot vector of length n_stations, and B_j is a one-hot diagonal matrix. So in this case, the sum is just the full
#  matrix, and the inverse is just the diagonal inverse. Let B = sum_j B^-1 and b=sum_j b_j be a diagonal matrix and vector. Then
#  the mean is
#        k^T (K + B)^-1 B B^-1 b
#       = k^T (K + B)^-1 b
#  as desired. 
#
# cov = d−k^T(K^-1 - K^-1 M K^-1)k^T


gp_adf1_alpha = []
gp_adf1_Kinv = []
for i in range(n_coefs):
    message_vars = np.array([B[i,i],] * len(train_stations))
    message_means = np.array([bs[j][i] for j in range(len(train_stations))])

    message_precs = 1.0/message_means
    
    K_noise_inv = np.linalg.inv(K_train + np.diag(message_vars))
    alpha = np.dot(K_noise_inv, message_means)
    gp_adf1_alpha.append(alpha)
    gp_adf1_Kinv.append(K_noise_inv)
    
    # should have 
    
    
    #print i, message_vars[0], K_noise_inv[0,0], message_means[0]
```

#### Training GPs with true ADF

```python
"""
How to do true ADF?

We maintain a block-diagonal K, which is to say, our assumed form is that the distribution over all coefs at all stations:
  - factors over coefs, so if we arrange by coef then the matrix is block-diagonal
  - *possibly furthermore*, all factors are the same except on the diagonal.
The way to represent this is as a cov matrix and a set of diagonal corrections, one for each coef.
"""

Kinv = np.linalg.inv(K_train)
adf_cov_invs = [Kinv.copy() for i in range(n_coefs)]
adf_means = [(np.zeros(n_train_stations,)) for i in range(n_coefs)]

full_prior = np.kron(np.eye(n_coefs), K_train)
full_prior_inv = np.linalg.inv(full_prior)
full_cov_inv = full_prior_inv.copy()
full_mean = np.zeros((full_prior.shape[0],))

approx_full_cov_inv = full_prior.copy()
approx_mean = full_mean.copy()

"""
At each step, we formally speaking
 - construct the full cov matrix over all coefs at all stations
 - update to include the current observation
 - find the nearest approximation
   - what is the nearest approximation? I assume we can show that the nearest block-diagonal approximation is just the obvious. 
     what if we require all off-diagonal terms to be constant?
     - if we require all *diagonal* terms to be constant when approximating a anisotropic diagonal Gaussian, we get something I just wrote on paper
     - we can do this in two steps:
        - first assume the best block-diag approximation is the obvious
        - then find the best identical-block-diag approximation to the block-diag
     - this seems hard. so let's instead just require block-diag, they don't have to be identical (though they might end up being so in this case?
- project this back down to the assumed form
"""

#let's make sure I'm doing the right thing here
# in the full update, we observe 
# z ~ (Aw, R)
#   ~ (AF.Tx, R) where F.T projects down from the full state to get coefs at this station
# we store the full state blocked by coef, so that we have
#  coef 1 station 1
#  coef 1 station 2
# ...
#  coef 2 station 1
#   ...
# and thus for station j, we want coef j, n_stations+j, 2*n_stations+j, etc.
# Now for a prior N(x; s, S)
# we multiply in z~N(AF'x, R)
# so we get a posterior N(x;P( S^-1s + FA'R^-1z)    P=(S^-1 + FA'R^-1AF')^-1)
# CHECK
# now for ADF:


F = np.zeros((full_prior.shape[0], 74))
for j in range(len(train_stations)):
    c = base_cs[j]
    Cc = cs[j]
        
    approx_mean_old = approx_mean.copy()
    approx_full_cov_inv[:,:]=0
    approx_mean[:] = 0
    for i in range(n_coefs):
        approx_full_cov_inv[i*n_train_stations:(i+1)*n_train_stations, i*n_train_stations:(i+1)*n_train_stations] = adf_cov_invs[i]
        approx_mean[i*n_train_stations:(i+1)*n_train_stations] = adf_means[i]
    assert( ((approx_mean_old-approx_mean) < 1e-8).all)
    
    F[:,:] = 0
    for i in range(n_coefs):
        F[i*n_train_stations+j, i] = 1
    
    i=8
    fci = full_cov_inv[i*n_train_stations:(i+1)*n_train_stations, i*n_train_stations:(i+1)*n_train_stations]
    afci = approx_full_cov_inv[i*n_train_stations:(i+1)*n_train_stations, i*n_train_stations:(i+1)*n_train_stations]
    #figure()
    #plot(np.diag(full_cov_inv))
    #plot(np.diag(approx_full_cov_inv))
    print "cov discrepancy", np.max(np.abs(fci-afci))
    #print "entries", np.max(np.abs(afci))
    
    # calculate true and approx posterior inverse covs
    inv_update = np.dot(F, np.dot(Cinv, F.T))
    full_cov_inv_new =  full_cov_inv + inv_update
    approx_cov_inv_new = approx_full_cov_inv + inv_update    
    
    # calculate updated means
    update_mean = np.dot(F, c)
    full_mean = np.linalg.solve(full_cov_inv_new, np.dot(full_cov_inv, full_mean) +  update_mean)
    approx_mean = np.linalg.solve(approx_cov_inv_new, np.dot(approx_full_cov_inv, approx_mean) +  update_mean)

    print "mean discrepancy", np.max(np.abs(full_mean-approx_mean))
    
    # now project the posterior back down to the ADF factorization
    approx_cov = np.linalg.inv(approx_cov_inv_new)
    for i in range(n_coefs):
        adf_cov_invs[i] = np.linalg.inv(approx_cov[i*n_train_stations:(i+1)*n_train_stations, i*n_train_stations:(i+1)*n_train_stations])
        adf_means[i] = approx_mean[i*n_train_stations:(i+1)*n_train_stations]

    full_cov_inv = full_cov_inv_new
    
    figure()
    plot(approx_mean)
    plot(full_mean)
```

```python
def predict_coef(loc, alpha, Kinv):
    ds = np.array([np.abs(loc-station_locations[j]) for j in train_stations])
    ks = w_var * np.exp(-.5*ds**2/lengthscale**2)
    return np.dot(ks, alpha), w_var - np.dot(ks, np.dot(Kinv, ks))

def predict_from_posterior(loc, ms, Minvs):
    # generally, given a Gaussian posterior, m, M we do GP prediction as
    # mean = k^T K^-1 m
    # cov = d−k^T(K^-1 - K^-1 M K^-1)k^T
    ds = np.array([np.abs(loc-station_locations[j]) for j in train_stations])
    ks = w_var * np.exp(-.5*ds**2/lengthscale**2)
    w = np.zeros((n_coefs,))
    v = np.zeros((n_coefs,))
    for i in range(n_coefs):
        w[i] = np.dot(ks, np.dot(Kinv, ms[i]))
        M = np.linalg.inv(Minvs[i])
        tmp = Kinv - np.dot(Kinv, np.dot(M, Kinv))
        v[i] = w_var + noise_var - np.dot(ks, np.dot(tmp , ks))
    return w, v

def predict_full_coefs(loc):
    ds = np.array([np.abs(loc-station_locations[j]) for j in train_stations])
    ks = w_var * np.exp(-.5*ds**2/lengthscale**2)
    w = np.zeros((n_coefs,))

    # we have a full posterior dist on (train station, coef).
    # now to predict, we need the test/train kernel matrix.
    # for each test coef i and train coef k*n_train_stations+j (coef k at station j)
    # this is 0 if i != k and 
    KS= np.zeros((n_coefs, n_train_stations*n_coefs))
    for i in range(n_coefs):
        for j in range(n_train_stations):
            KS[i, i*n_train_stations+j] = ks[j]
            
    pred = np.dot(KS, np.dot(full_prior_inv, full_mean))
    M = np.linalg.inv(full_cov_inv)
    tmp = full_prior_inv - np.dot(full_prior_inv, np.dot(M, full_prior_inv))

    C = np.eye(n_coefs)*(w_var + noise_var) - np.dot(KS, np.dot(tmp , KS.T))
    return pred, C
    
        
def predict_wavelet(loc, alphas, Kinvs):
    ds = np.array([np.abs(loc-station_locations[j]) for j in train_stations])
    ks = w_var * np.exp(-.5*ds**2/lengthscale**2)
    
    w = np.zeros((n_coefs,))
    
    v = np.zeros((n_coefs,))
    for i in range(n_coefs):
        w[i] = np.dot(ks, alphas[i])
        v[i] = w_var + noise_var - np.dot(ks, np.dot(Kinvs[i], ks))
    return w, v

sum_factored = 0
sum_adf1 = 0
sum_kalman = 0
sum_adf = 0
sum_full = 0
prior_sum = 0
for j,s in enumerate(station_locations):
    w = ws_authentic[j]
    Aw = np.dot(A, w)
    #min1, max1 = np.min(zs[j]), np.max(zs[j])
    #min2, max2 = np.min(Aw), np.max(Aw)
    #fmin, fmax = min(min1, min2), max(max1, max2)
    fmin,fmax=-4,4
    
    f = plt.figure(figsize=(20,5))
    ax1 = f.add_subplot(121)
    ax1.plot(zs[j])
    ax1.set_ylim((fmin, fmax))
    ax1.set_title("observed signal at x=%.2f (train=%s)" % (s, ["%.2f" % station_locations[j] for j in train_stations]))
    ax2 = f.add_subplot(122)
    
    ax2.plot(np.dot(A, w), label='true')
    w_pred_factored, w_var_factored = predict_wavelet(s, gp_alpha, gp_Kinv)
    w_pred_adf1, w_var_adf1 = predict_wavelet(s, gp_adf1_alpha, gp_adf1_Kinv)
    w_pred_kalman, w_var_kalman = predict_wavelet(s, kalman_adf1_alpha, kalman_adf1_Kinv)
    w_pred_adf, w_var_adf = predict_from_posterior(s, adf_means, adf_cov_invs)
    w_pred_full, w_cov_full = predict_full_coefs(s)
    
    if np.isnan(w_pred_adf).any() or np.isnan(w_var_adf.any()) or (w_var_adf <= 0).any():
        import pdb; pdb.set_trace()


    prior_likelihood = np.mean([scipy.stats.norm.logpdf(w[i], loc=0, scale=np.sqrt(w_var)) for i in range(n_coefs)])

    factored_likelihood = np.mean([scipy.stats.norm.logpdf(w[i], loc=w_pred_factored[i], scale=np.sqrt(w_var_factored[i])) for i in range(n_coefs)])
    adf1_likelihood = np.mean([scipy.stats.norm.logpdf(w[i], loc=w_pred_adf1[i], scale=np.sqrt(w_var_adf1[i])) for i in range(n_coefs)])
    kalman_likelihood = np.mean([scipy.stats.norm.logpdf(w[i], loc=w_pred_kalman[i], scale=np.sqrt(w_var_kalman[i])) for i in range(n_coefs)])
    adf_likelihood = np.mean([scipy.stats.norm.logpdf(w[i], loc=w_pred_adf[i], scale=np.sqrt(w_var_adf[i])) for i in range(n_coefs)])
    full_likelihood = scipy.stats.multivariate_normal.logpdf(w, mean=w_pred_full, cov=w_cov_full)/n_coefs

    ax2.plot(np.dot(A, w_pred_factored), label="factored %.3f" % (factored_likelihood))
    ax2.plot(np.dot(A, w_pred_adf1), label="fakeadf %.3f" % (adf1_likelihood))
    ax2.plot(np.dot(A, w_pred_kalman), label="kalman %.3f" % (kalman_likelihood))
    ax2.plot(np.dot(A, w_pred_adf), label="adf %.3f" % (adf_likelihood))
    ax2.plot(np.dot(A, w_pred_full), label="full %.3f" % (full_likelihood))
    ax2.set_ylim((fmin, fmax))
    
    prior_sum += prior_likelihood
    sum_factored += factored_likelihood
    sum_adf1 += adf1_likelihood
    sum_kalman += kalman_likelihood
    sum_adf += adf_likelihood
    sum_full += full_likelihood
    ax2.set_title("inferred wave: prior lik %.3f" % (prior_likelihood))
    
    ax2.legend()
    
print "mean log-likelihoods:"
print "prior", prior_sum/len(station_locations)
print "factored", sum_factored/len(station_locations)
print "kalman", sum_kalman/len(station_locations)
print "fakeadf", sum_adf1/len(station_locations)
print "adf", sum_adf/len(station_locations)
print "full",sum_full/len(station_locations)
```
