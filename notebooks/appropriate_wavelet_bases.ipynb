```python
from sigvisa.ssms_c import CompactSupportSSM
import pyublas 

def get_cssm(basis, coef_mean, coef_var, obs_var=0.01):
    sts, ets, ids, prototypes, N = basis
    starray = np.array(sts, dtype=np.int32)
    etarray = np.array(ets, dtype=np.int32)
    idarray = np.array(ids, dtype=np.int32)
    n1 = len(prototypes)
    n2 = np.max([len(l) for l in prototypes])
    print n1, n2
    m = np.matrix(np.ones((n1, n2), dtype=np.float64)*np.nan)
    for i, p in enumerate(prototypes):
        m[i,:len(p)] = p
    
    n = len(starray)
    coef_means = np.array(np.ones((n,))*coef_mean, dtype=np.float)
    coef_vars = np.array(np.ones((n,))*coef_var, dtype=np.float)
        
    cssm = CompactSupportSSM(starray, etarray, idarray, m, coef_means, coef_vars, obs_var, 0.0)
    print "built cssm with max dimension", cssm.max_dimension()
    return cssm

def get_cssm_py(basis, coef_mean, coef_var):
    import sigvisa.models.statespace.compact_support as cs
    sts, ets, ids, prototypes, N = basis
    
    n = len(sts)
    coef_means = np.array(np.ones((n,))*coef_mean, dtype=np.float)
    coef_vars = np.array(np.ones((n,))*coef_var, dtype=np.float)
    cssm = cs.ImplicitCompactSupportSSM(sts, ets, ids, prototypes, coef_means, coef_vars, 0.0, 0.01)
    return cssm

def visualize_wavelet_basis(basis, len_s, coef_mean=0.0, coef_var=1.0, samples=2):
    cssm = get_cssm(basis, coef_mean=coef_mean, coef_var=coef_var)
    N = np.max(basis[0])
    axes = visualize_ssm(cssm, len_s, N, samples=samples)
    return axes, cssm

def visualize_ssm(ssm, len_s, N, samples=2):
    m = ssm.mean_obs(N)
    v = ssm.obs_var(N)
    samples = [ssm.prior_sample(N) for i in range(samples)]
    
    fig = plt.figure(figsize=(10, 5), dpi=144)
    fig.patch.set_facecolor('white')
    axes = fig.add_subplot(111)
    axes.set_xlabel("Time (s)", fontsize=8)

    x = np.linspace(0, len_s, N)

    axes.plot(x, m, c="black", lw=2)
    axes.plot(x, m + 2*np.sqrt(v), c="red", lw=2)
    axes.plot(x, m - 2*np.sqrt(v), c="red", lw=2)
    for s in samples:
        axes.plot(x, s)
        
    return axes

```
