```python
from sigvisa import Sigvisa

# we want to be able to say, for each phase, "conditioned on phase A, what are the probabilities of seeing phases X,Y,Z,..."
# so we want, first, a list of event station-pairs, containing the phases observed for each

s = Sigvisa()
cursor = s.dbconn.cursor()
sql_query = "select lebo.evid, s.sta, leba.phase from leb_origin lebo, leb_assoc leba, static_siteid s where lebo.orid=leba.orid and leba.sta=s.sta"
cursor.execute(sql_query)
arrivals = cursor.fetchall()

from collections import defaultdict

phasesets = defaultdict(set)
for (evid, sta, phase) in arrivals:
    phasesets[(evid, sta)].add(phase)
```

```python
from collections import Counter

def phase_conditional(phase):
    counts = Counter()
    for phaseset in phasesets.values():
        if phase in phaseset:
            counts.update(phaseset)
    total = counts[phase]
    
    sorted_keys = sorted(counts.keys(), key = lambda k: -counts[k])[1:]
    sorted_probs = [float(counts[key])/total for key in sorted_keys]
    
    figure(figsize=(20, 4))
    title("%s %d" % (phase, total))
    plt.bar(range(len(sorted_probs)), sorted_probs, align='center')
    plt.xticks(range(len(sorted_probs)), sorted_keys)
    
phase_conditional('P')
phase_conditional('Pn')
phase_conditional('Pg')
phase_conditional('S')
phase_conditional('Sn')
phase_conditional('LR')
phase_conditional('Lg')
```
