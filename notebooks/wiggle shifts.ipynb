```python
from sigvisa.models.wiggles.fourier_features import FourierFeatureGenerator
from sigvisa.models.wiggles import load_wiggle_generator
```

```python
wg = load_wiggle_generator(94)
```

```python
x = np.linspace(0, 60, 300)
sinx = np.sin(x) + 1
plot(x,sinx)
```

```python
features = wg.features_from_signal(sinx)
```

```python
print features
```

```python
plot(x,wg.signal_from_features(features))
```

```python
import copy
a = wg.param_dict_to_array(features)
b = np.array(a, copy=True)
c = np.array(a, copy=True)
wg.timeshift_param_array(a,4.0)
wg.timeshift_param_array(b,8.0)
wg.timeshift_param_array(c,16.0)
s1 = wg.signal_from_features(a)
s0 = wg.signal_from_features(features)
idx = 20
plot(x, s0, c='green')
plot(x[idx:], s1[:-idx], c='red')
```

```python
farray = np.random.randn(96)


a = copy.copy(farray)
wg.timeshift_param_array(a,4.0)
s1 = wg.signal_from_features(a)
s0 = wg.signal_from_features(farray)
idx = 20
plot(x, s0, c='green')
plot(x[idx:], s1[:-idx], c='red')
```
