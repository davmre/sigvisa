```python
from sigvisa.models.distributions import Laplacian
from sigvisa.models.ttime import tt_predict
from sigvisa.source.event import Event
import seaborn as sns
from sigvisa import Sigvisa
```

```python
s = Sigvisa()
slon, slat = s.earthmodel.site_info("FITZ", 0)[0:2]
print slon, slat
```

```python
tts = []
for i in range(300):
    d1 = np.random.rand()
    d2 = np.random.rand()
    lon = -130+d1
    lat = -16+d2
    ev = Event(lon=lon, lat=lat, depth=0, time=1238890000, mb=4.0)
    tt_mean = tt_predict(ev, "FITZ", "S")
    tt = Laplacian(tt_mean, 3.0).sample()
    tts.append(tt)
    
tts = np.array(tts, dtype=np.float)
```

```python
sns.distplot(tts)
xs = np.linspace(np.min(tts), np.max(tts), 100)
l = Laplacian(np.mean(tts), 10.0)
ps = [np.exp(l.log_p(x)) for x in xs]
plot(xs, ps)
```
