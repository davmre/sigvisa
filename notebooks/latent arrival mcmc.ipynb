```python
from sigvisa import Sigvisa
from sigvisa.signals.io import load_event_station_chan
from sigvisa.graph.sigvisa_graph import SigvisaGraph
from sigvisa.source.event import get_event


s = Sigvisa()


tmtype = {'amp_transfer': 'param_sin1', 'tt_residual': 'constant_laplacian', 'coda_decay': 'param_linear_distmb', 'peak_offset': 'param_linear_mb', 'peak_decay': 'dummy'}

def load_sg(evid=5303023, sta='MKAR', chan='BHZ'):
    sg = SigvisaGraph(template_model_type='dummy', template_shape='paired_exp', wiggle_model_type='dummy', wiggle_family='dummy',
                      phases=['P'], nm_type = 'ar', run_name='everything', iteration=2, fixed_arrival_npts=10000)
    
    ev = get_event(evid=evid)
    
    
    cursor = s.dbconn.cursor()
    wave = load_event_station_chan(evid, sta, chan, cursor=cursor).filter("%s;env" % 'freq_2.0_3.0')
    
    cursor.close()
    
    wn = sg.add_wave(wave=wave, fixed=True)
    
    #from sigvisa.models.noise.armodel.model import ARModel, ErrorModel
    #em = ErrorModel(mean=0, std=0.00054688)
    #ar = ARModel(params=np.array([.6, .1]), c=0.072982422, em=em)
    #wn.nm = ar
    
    sg.add_event(ev=ev, fixed=True)
    
    latent = [n for (k,n) in wn.parents.items() if 'latent_arrival' in k][0]
    
    np.random.seed(0)
    sg.parent_sample_all()
    return sg, wn, latent

#import cPickle as pickle
#with open('gibbs.sg', 'r') as f:
#    sg = pickle.load(f)
#    wn = sg.station_waves['FIA0'][0]

def debug_plot_all(sg, latent, wn, msg="", i=None):
    if msg =="" and i is not None:
        msg = "round %d" % i
    figure()
    wn.debugging_plot(ax=gca(), plot_mode="full")
    title('%s: observed signal, overall lp %f' % (msg,sg.current_log_p()))
    figure()
    latent.debugging_plot(ax=gca(), plot_mode="predict")
    title('%s: latent signal with predicted shape' % (msg,))
    figure()
    latent.debugging_plot(ax=gca(), plot_mode="wiggle")
    title('%s: wiggle lp %f' % (msg,latent.log_p()))
    figure()
    wn.debugging_plot(ax=gca(), plot_mode="noise")
    title('%s: noise lp %f' % (msg,wn.log_p()))
   

my_sg, my_wn, my_latent = load_sg()
debug_plot_all(my_sg, my_latent, my_wn)
```
