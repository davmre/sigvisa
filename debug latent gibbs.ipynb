```python
from sigvisa import Sigvisa
from sigvisa.signals.common import Waveform
from sigvisa.graph.sigvisa_graph import SigvisaGraph
from sigvisa.models.noise.armodel.model import ARModel, ErrorModel
from sigvisa.source.event import get_event


def debug_plot_all(sg, latent, wn, msg="", i=None):
    if msg =="" and i is not None:
        msg = "round %d" % i
    figure()
    wn.debugging_plot(ax=gca(), plot_mode="full")
    title('%s: observed signal, overall lp %f' % (msg,sg.current_log_p()))
    figure()
    latent.debugging_plot(ax=gca(), plot_mode="predict")
    title('%s: latent signal with predicted shape' % (msg,))
    figure()

    latent.debugging_plot(ax=gca(), plot_mode="wiggle")
    title('%s: wiggle lp %f' % (msg,latent.log_p()))
    figure()
    wn.debugging_plot(ax=gca(), plot_mode="noise")
    title('%s: noise lp %f' % (msg,wn.log_p()))

tmtype = {'amp_transfer': 'param_sin1', 'tt_residual': 'constant_laplacian', 'coda_decay': 'param_linear_distmb', 'peak_offset': 'param_linear_mb'}
sg = SigvisaGraph(template_model_type='dummy', template_shape='paired_exp', wiggle_model_type='dummy', wiggle_family='dummy',
                    phases=['P'], nm_type = 'ar', run_name='everything', iteration=2)

obs_data = np.zeros((1,))
#obs_data = np.array((.33, 4.0, 7.0, 3.5, 1.8, 1.0, .7, .5, .4, .36))
obs_wave = Waveform(data=obs_data, stime=1238982080.0, srate=1, sta="MKAR", chan="BHZ", filter_str="freq_2.0_3.0")
wn = sg.add_wave(obs_wave)
ev = get_event(evid=5303023)
sg.add_event(ev=ev, fixed=True)
latent = [n for (k,n) in wn.parents.items() if 'latent_arrival' in k][0]
np.random.seed(0)

fake_em = ErrorModel(mean=0, std=0.01)
fake_noise = ARModel(params=(0.5,), c=0.0, em=fake_em)
wn.nm = fake_noise

fake_em2 = ErrorModel(mean=0, std=0.01)
fake_wiggle = ARModel(params=(0.5,), c=0.0, em=fake_em2)
latent.arwm = fake_wiggle


sg.parent_predict_all()


tmnodes = sg.get_arrival_nodes(1, 'MKAR', 'P', 'freq_2.0_3.0', 'BHZ')
k_decay, n_decay = tmnodes['coda_decay']
k_height, n_height = tmnodes['coda_height']
k_offset, n_offset = tmnodes['peak_offset']

print n_offset.get_value(), n_height.get_value(), n_decay.get_value()

#n_decay.set_value(0.0)
#n_height.set_value(2.0)
#n_offset.set_value(1.0)
n_decay.set_value(0.0)
n_height.set_value(1.0)
n_offset.set_value(1.0)
latent.parent_predict()
print latent.npts
#latent.set_value(np.array((.02, 4.0, 7.0, 3.5, 1.8, 1.0, .7, .5, .4, .36)))
latent.set_value(np.array((.02, )), force_new_size=True)

latent.npts = 1
latent._tmpl_shape_cache = latent._tmpl_shape_cache[:1]
debug_plot_all(sg, latent, wn)
```
