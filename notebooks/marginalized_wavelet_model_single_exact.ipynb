```python
def dwt(x, len_s, wtype='db9'):
    cc = unflatten_wavelet_coefs(x, wtype=wtype, len_s=len_s)
    return pywt.waverec(cc, wtype, 'zpd')
    
def idwt(s, wtype='db9', levels=999):
    return flatten_wavelet_coefs(pywt.wavedec(s, wtype, 'zpd'), levels=levels)
```

```python
def innovation_cov(P, len_s, wtype):
    # P is a wxw matrix giving the covariance on the wavelet params
    # want to return HPH' where H is the dwt matrix
    # = H(HP)'
    
    k = P.shape[0]
    
    HP = np.empty((len_s, k))
    
    # enumerating P.T gives rows of P.T, which are cols of P
    for i, col in enumerate(P.T):
        HP[:, i] = dwt(col, len_s, wtype)
    
    # now we want cols of (HP)'
    # which are rows of HP
    HPHt = np.empty((len_s, len_s))
    for i, row in enumerate(HP):
        HPHt[:, i] = dwt(row, len_s, wtype)
    
    return HPHt

```

```python
def signal_cov(w_cov, w_len_idx, ar_mult, ar_noise, template):
    n = len(template)
    Sx = ar_mult.cov_matrix(n)
    Sz = ar_noise.cov_matrix(n)
    
    W = Sx.copy()
    
    AEAt = innovation_cov(w_cov, w_len_idx, 'db9')
    W[:w_len_idx, :w_len_idx] += AEAt
    
    for i in range(n):
        W[:,i] *= template
        W[i,:] *= template
        
    return W + Sz

def signal_mean(w_mean, w_len_idx, template):
    pred_rw = dwt(w_mean, w_len_idx, 'db9')
    pred_wiggle = np.ones((len(template),))
    pred_wiggle[:len(pred_rw)] += pred_rw
    
    pred_signal = template * pred_wiggle
    
    return pred_signal

```

```python
weird_prior_vars = np.arange(1,75)
wssm = CompactSupportSSM(small_bases, bias=0.0, coef_prior_means=prior_means, coef_prior_vars=weird_prior_vars)
cmeans, cvars = wssm.filtered_coef_marginals(z)

# we have a gaussian with mean cmeans, cvars
# let's divide by a prior density. 
# if we have a Gaussian N(a, A) and multiply by N(b, B), we get C = 1/(1/A + 1/B) and c = (a/A + b/B)/(1/A+1/B)

def multiply_gaussian(m1, v1, m2, v2):
    cv = 1.0/(1.0/v1 + 1.0/v2)
    mv = cv * (m1/v1 + m2/v2)
    mv = np.where(v1==0, m1, mv)
    return mv, cv

subtracted_means1, subtracted_vars1 = multiply_gaussian(cmeans, cvars, prior_means, -weird_prior_vars)

weird_prior_vars = np.arange(1,75)[::-1]
wssm = CompactSupportSSM(small_bases, bias=0.0, coef_prior_means=prior_means, coef_prior_vars=weird_prior_vars)
cmeans, cvars = wssm.filtered_coef_marginals(z)
subtracted_means2, subtracted_vars2 = multiply_gaussian(cmeans, cvars, prior_means, -weird_prior_vars)

subtracted_flat_means, subtracted_flat_vars = multiply_gaussian(flat_means, flat_vars, prior_means, -prior_vars)
```

```python
def multiply_mvn(a, A, b, B, P=None):
    Ainv = np.linalg.inv(A)
    Binv = np.linalg.inv(B)
    if P is None:
        C = np.linalg.inv(Ainv+Binv)
        c = np.dot(C, np.dot(Ainv, a) + np.dot(Binv, b))
    else:
        C = np.linalg.inv(Ainv+np.dot(P.T, np.dot(Binv, P)))
        c = np.dot(C, np.dot(Ainv, a) + np.dot(P.T, np.dot(Binv, b)))
    return c, C

v = 1.3
noise = 1.0
obs = 2.7

posterior_C = np.linalg.inv(np.eye(2)/v + np.ones((2,2))/noise)
posterior_c = np.dot(posterior_C, np.ones((2,))*obs/noise)
print posterior_c
print posterior_C

prior_c = np.zeros((2,))
prior_C = np.eye(2)*v

print multiply_mvn(posterior_c, posterior_C,prior_c, -prior_C)
```
