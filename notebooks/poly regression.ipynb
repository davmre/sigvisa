```python
from sigvisa.models.spatial_regression.baseline_models import LinearBasisModel

k = 3
def poly_basisfns(order):
    basisfn_strs = ["lambda x : " + ("1" if d==0 else "x**%d" % d)   for d in range(order+1)]
    return [eval(s) for s in basisfn_strs]
my_bfns = poly_basisfns(k)
my_b = np.zeros((k+1,))
my_B = np.eye(k+1) * 9
my_H = np.array([[f1(x) for f1 in my_bfns] for x in X])
lbm = LinearBasisModel(X=X, y=y, basisfns=my_bfns, param_mean=my_b, param_covar=my_B, noise_std=sigma_n)
a = lbm.predict(X)
plot(X, a, color='red')

print X.shape
print f.shape
plot(X, f, color='blue')

param_covar = dot(lbm.sqrt_covar.T, lbm.sqrt_covar)
for i in range(10):
    p = np.random.multivariate_normal(mean=lbm.mean, cov=param_covar)
    a = np.dot(my_H, p) #lbm.sample(X)
    plot(X, a)
    
print lbm.ll
```
