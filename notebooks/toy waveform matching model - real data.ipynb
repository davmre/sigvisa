```python
# PLOT: signals from both events
#       separate and superimposed wavelet features 
#       reconstructed wavelet signal (by itself and over true signal)
#       do all of this with proper timeline labeling, etc


def coef_pyramid_plot(coefs, first=0, scale='uniform', ax=None, **kwargs):
    """
    Parameters
    ----------
    coefs : array-like
        Wavelet Coefficients. Expects an iterable in order Cdn, Cdn-1, ...,
        Cd1, Cd0.
    first : int, optional
        The first level to plot.
    scale : str {'uniform', 'level'}, optional
        Scale the coefficients using the same scale or independently by
        level.
    ax : Axes, optional
        Matplotlib Axes instance

    Returns
    -------
    Figure : Matplotlib figure instance
        Either the parent figure of `ax` or a new pyplot.Figure instance if
        `ax` is None.
    """

    if ax is None:
        import matplotlib.pyplot as plt
        fig = plt.figure()
        ax = fig.add_subplot(111, axisbg='lightgrey')
    else:
        fig = ax.figure

    n_levels = len(coefs)
    n = 2**(n_levels - 1) # assumes periodic

    if scale == 'uniform':
        biggest = [np.max(np.abs(np.hstack(coefs)))] * n_levels
    else:
        # multiply by 2 so the highest bars only take up .5
        biggest = [np.max(np.abs(i))*2 for i in coefs]

    for i in range(first,n_levels):
        x = np.linspace(2**(n_levels - 2 - i), n - 2**(n_levels - 2 - i), len(coefs[i]))
        ymin = n_levels - i - 1 + first
        yheight = coefs[i]/biggest[i]
        ymax = yheight + ymin
        print first, n_levels, i, x.shape, ymin, ymax.shape
        ax.vlines(x, ymin, ymax, **kwargs)

    ax.set_xlim(0,n)
    ax.set_ylim(first - 1, n_levels)
    ax.yaxis.set_ticks(np.arange(n_levels-1,first-1,-1))
    ax.yaxis.set_ticklabels(np.arange(first,n_levels))
    ax.xaxis.set_ticks([])
    ax.tick_params(top=False, right=False, direction='out', pad=6)
    ax.set_ylabel("Levels", fontsize=14)
    ax.grid(True, alpha=.85, color='white', axis='y', linestyle='-')
    #ax.set_title('Wavelet Detail Coefficients', fontsize=16,
    #        position=(.5,1.05))
    fig.subplots_adjust(top=.89)

    return fig

sta='AKTO'
#figure(figsize=(15,4))

matplotlib.rcParams['axes.grid'] = False
matplotlib.rcParams['axes.facecolor'] = 'white'

c1 = unflatten_wavelet_coefs(wavelets_evid(evs[sta], ys[sta], 4760683), 'db9')
c2 = unflatten_wavelet_coefs(wavelets_evid(evs[sta], ys[sta], 5891632), 'db9')

c3 = unflatten_wavelet_coefs(wavelets_evid(evs[sta], ys[sta], 4889362), 'db9')


fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(111)
ax.patch.set_alpha(0.0)
f = coef_pyramid_plot(c1[:4], scale=True, ax=ax, color='blue', linewidth=8)

fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(111)
ax.patch.set_alpha(0.0)
f = coef_pyramid_plot(c2[:4], scale=True, ax=ax, color='green', linewidth=8)

fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(111)
ax.patch.set_alpha(0.0)
f = coef_pyramid_plot(c3[:4], scale=True, ax=ax, color='red', linewidth=8)


fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(111)

f = coef_pyramid_plot(c1[:4], scale=True, ax=ax, color='blue', linewidth=10, alpha=0.8)
f = coef_pyramid_plot(c2[:4], scale=True, ax=ax, color='green', linewidth=8, alpha=0.8)

fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(111)

f = coef_pyramid_plot(c1[:4], scale=True, ax=ax, color='blue', linewidth=10, alpha=0.8)
f = coef_pyramid_plot(c3[:4], scale=True, ax=ax, color='red', linewidth=8, alpha=0.8)


#plot(wavelets_evid(evs[sta], ys[sta], 5891632))

matplotlib.rcParams['axes.grid'] = False
matplotlib.rcParams['axes.facecolor'] = 'white'
#matplotlib.rcParams['axes.facealpha'] = 0.0


figure(figsize=(15,7), frameon=False, facecolor='white')
xx = np.linspace(-10, 50, 1200)
evX, n = get_ev_X(evs[sta], evid=4760683)
plot(xx, signals[sta][n,:], linewidth=2)
evX, n = get_ev_X(evs[sta], evid=5891632)
plot(xx, signals[sta][n,:])
gca().yaxis.set_ticks([])
gca().patch.set_alpha(0.0)

figure(figsize=(15,7), frameon=False, facecolor='white')
xx = np.linspace(-10, 50, 1200)
evX, n = get_ev_X(evs[sta], evid=4889362)
plot(xx, signals[sta][n,:], color='red', alpha=0.8)
evX, n = get_ev_X(evs[sta], evid=4760683)
plot(xx, signals[sta][n,:], linewidth=2, color='blue')
gca().yaxis.set_ticks([])
gca().patch.set_alpha(0.0)

figure(figsize=(15,7), frameon=False, facecolor='white')
xx = np.linspace(0, 30, 600)
evX, n = get_ev_X(evs[sta], evid=4760683)
plot(xx, signals[sta][n,:][213:813], linewidth=2)
evX, n = get_ev_X(evs[sta], evid=5891632)
plot(xx, signals[sta][n,:][213:813])
gca().yaxis.set_ticks([])
gca().patch.set_alpha(0.0)


figure(figsize=(15,7), frameon=False, facecolor='white')
xx = np.linspace(0, 30, 600)
evX, n = get_ev_X(evs[sta], evid=4889362)
plot(xx, signals[sta][n,200:800], color='red', alpha=0.8)
evX, n = get_ev_X(evs[sta], evid=4760683)
plot(xx, signals[sta][n,200:800], linewidth=2, color='blue')
gca().yaxis.set_ticks([])
gca().patch.set_alpha(0.0)
    
figure(figsize=(15,7), frameon=False, facecolor=None)
xx = np.linspace(0, 30, 600)
plot(xx, pywt.waverec(c1, 'db9', 'per'), linewidth=3, alpha=.8)
plot(xx, pywt.waverec(c2, 'db9', 'per'), linewidth=2, alpha=.8)
gca().yaxis.set_ticks([])
gca().patch.set_alpha(0.0)


figure(figsize=(15,7), frameon=False, facecolor=None)
xx = np.linspace(-10, 50, 1200)
plot(xx, signals[sta][n,:]/np.linalg.norm(signals[sta][n,:]), linewidth=4, alpha=.8, color='blue')
xx = np.linspace(0.65, 30.65, 600)
plot(xx, pywt.waverec(c2, 'db9', 'per'), linewidth=3, alpha=.8, color='purple')
gca().yaxis.set_ticks([])
gca().patch.set_alpha(0.0)
```

```python
# need to generate:
# a test event object
# a ys[] object with training data (but NOT the test event)
# a ys[] object with the test event: just picked arrival times and 'loaded' arrival times 
# (this will be a new field: giving the times the loaded signals are aligned to. for synth datak this isthe true time; for real data, it can be the pred atime)

def isolate_test_ev(evs, ys, evid):
    new_evs = dict()
    new_ys = dict()
    
    test_ys = dict()
    ats_to_load = dict()
    ev = None
        
    for sta in ys.keys():
        evid_match = (evs[sta][:, 4] == evid) 
        if np.sum(evid_match) == 1:
            x_ev = evs[sta][evid_match,:]
            ev = Event(lon=x_ev[0,0], lat=x_ev[0,1], depth=0, time=x_ev[0,3], mb=x_ev[0,5])
            print "got ev", ev
            break
            
    for sta in ys.keys():
        new_ys[sta] = dict()
        test_ys[sta] = dict()
        
        evid_match = (evs[sta][:, 4] == evid)        
        ev_n = np.arange(len(evid_match))[evid_match]
        if len(ev_n) == 0:
            print "event not observed at", sta
            new_evs[sta] = evs[sta]
            new_ys[sta] = ys[sta]
        elif len(ev_n) == 1:
            print "event observed as %d at %s" % (ev_n, sta)
            new_evs[sta] = evs[sta][~evid_match,:]
            for k in ys[sta].keys():
                new_ys[sta][k] = ys[sta][k][~evid_match]
            test_ys[sta]['arrival_time_picked'] = ys[sta]['arrival_time_picked'][evid_match]
            
        else:
            raise Exception('multiple events at %s match evid %d' % (sta, evid))

        try:
            pred_at = ev.time + tt_predict(ev, sta, phase)
            ats_to_load[sta] = pred_at
            test_ys[sta]['arrival_time_loaded'] = pred_at
        except:
            print "can't predict travel time to %s" % sta
        
        
    return new_evs, new_ys, test_ys, ats_to_load, ev

test_evid = 5891632
doublet_evid = 4760683
train_evs, train_ys, test_ys, ats_to_load, test_ev = isolate_test_ev(evs, ys, test_evid)
```
