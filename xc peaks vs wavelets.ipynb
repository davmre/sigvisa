```python
from sigvisa.explore.doublets.xcorr_pairs import xcorr, extract_phase_window
from sigvisa.utils.geog import dist_km
from sigvisa import Sigvisa
import seaborn as sns
import pandas as pd
import pywt

def flatten_wavelet_coefs(coefs):
    return np.concatenate((coefs[0], coefs[1], coefs[2]))

def unflatten_wavelet_coefs(cc):    
    coefs = [cc[:19], cc[19:38], cc[38:76], np.zeros((75,)), np.zeros((150,)), np.zeros((300,))]
    return coefs

def get_leb_atime(cursor, evid, sta, phase='P'):
    sql_query = "select l.time from leb_arrival l, leb_assoc leba, leb_origin lebo where lebo.evid=%d and leba.phase='%s' and l.sta='%s' and leba.orid=lebo.orid and l.arid=leba.arid" % (evid, phase, sta)
    cursor.execute(sql_query)
    try:
        r = cursor.fetchall()[0]
        return float(r[0])
    except:
        return np.nan
    
    
def load_signals(sta, X):
    signals = np.zeros((X.shape[0], 800))
    s = Sigvisa()
    cursor = s.dbconn.cursor()
    
    for (i, x) in enumerate(X):
        evid = int(x[4])
        try:
            atime = get_leb_atime(cursor, evid, sta)
            if np.isnan(atime): continue
            w, srate = extract_phase_window(sta, 'auto', 'P', atime, 15.0, 'freq_0.8_3.5', evid)
            signals[i,:] = w
        except Exception as e:
            print "error at evid %d: %s" % (evid, e)
            continue
    cursor.close()
    return signals

def signals_to_wavelets(S):
    wavelets = np.zeros((S.shape[0], 76))
    for (i, s) in enumerate(S):
        if np.std(s) == 0: continue
        s = (s - np.mean(s)) / (np.std(s) * np.sqrt(len(s)))        
        coefs = pywt.wavedec(s[200:800], 'db8', mode='per')
        wavelets[i,:] = flatten_wavelet_coefs(coefs)
    return wavelets
        
def xcorr_pairs(S1, S2):
    xcorrs = np.zeros((S1.shape[0], S2.shape[0]))
    for (i, s1) in enumerate(S1):
        for (j, s2) in enumerate(S2):
            xcmax, offset = xcorr(s1, s2)
            if np.isnan(xcmax): continue
            xcorrs[i,j] = xcmax
    return xcorrs

def distance_pairs(X1, X2):
    dists = np.zeros((X1.shape[0], X2.shape[0]))
    for (i, x1) in enumerate(X1):
        for (j, x2) in enumerate(X2):
            dists[i,j] = dist_km((x1[0], x1[1]), (x2[0], x2[1]))
    return dists

def plot_aligned(s1, s2):
    xcmax, offset = xcorr(s1, s2)
    x = np.linspace(-5, 15, 800)
    plot(x, s1/np.linalg.norm(s1, 2))
    plot(x + offset / 40.0, s2/np.linalg.norm(s2, 2))
    
def wavelet_pairs(W1, W2, coef_start=0, coef_end=None):
    if coef_end is None:
        coef_end = W1.shape[1]
    dots =  np.zeros((W1.shape[0], W2.shape[0]))
    for (i, w1) in enumerate(W1):
        for (j, w2) in enumerate(W2):
            n1 = w1[coef_start:coef_end] 
            n1 = n1 - np.mean(n1)
            n1 = n1 / np.linalg.norm(n1, 2)
            
            n2 = w2[coef_start:coef_end] 
            n2 = n2 - np.mean(n2)
            n2 = n2 / np.linalg.norm(n2, 2)            
            
            dots[i,j] = np.dot(n1, n2)
            
    return dots
```

```python
stas = ['MKAR', 'TORD', 'ASF', 'MMAI', 'KEST', 'VRAC', 'AKASG', 'EIL', 'BRTR', 'GERES']

d = pd.DataFrame()

def cache_npy(fname, f, *args, **kwargs):
    try:
        return np.load(fname)
    except IOError:
        r = f(*args, **kwargs)
        np.save(fname, r)
        return r
    
for sta in stas:
    print sta

    signals_train = cache_npy('signals_%s.npy' % sta, load_signals, sta, X_train)
    if (np.sum(np.abs(signals_train))) < 1e-8:
        print "CACHE ERROR"
        import pdb; pdb.set_trace()
    xcorr_train_train = cache_npy('xcorr_%s.npy' % sta, xcorr_pairs, signals_train, signals_train)
    dist_train_train = cache_npy('dists_%s.npy' % sta, distance_pairs, X_train, X_train)
    w_train = cache_npy('wavelets_%s.npy' % sta, signals_to_wavelets, signals_train)

    
    wavelet_dots = wavelet_pairs(w_train, w_train)
    wavelet_dots_19 = wavelet_pairs(w_train, w_train, coef_end=19)
    wavelet_dots_38 = wavelet_pairs(w_train, w_train, coef_end=38)
    wavelet_dots_1938 = wavelet_pairs(w_train, w_train, coef_start=19, coef_end=38)
    wavelet_dots_3876 = wavelet_pairs(w_train, w_train, coef_start=38, coef_end=76)

    
    
    print dist_train_train.shape
    print xcorr_train_train.shape
    d['dist_%s' % sta] = dist_train_train.flatten()
    d['xcorr_%s' % sta] = xcorr_train_train.flatten()
    d['wdot_%s' % sta] = wavelet_dots.flatten()
    d['wdot_abs_%s' % sta] = np.abs(wavelet_dots.flatten())
    d['wdot19_abs_%s' % sta] = np.abs(wavelet_dots_19.flatten())
    d['wdot38_abs_%s' % sta] = np.abs(wavelet_dots_38.flatten())
    d['wdot1938_abs_%s' % sta] = np.abs(wavelet_dots_1938.flatten())
    d['wdot3876_abs_%s' % sta] = np.abs(wavelet_dots_3876.flatten())

    d.to_csv('d_tmp_%s.csv' % sta)
d = d.replace(0, np.nan)
self_indices = np.arange(0, d.shape[0], 193)
#d.iloc[self_indices,:] = np.nan
d.to_csv('d_final.csv')
```
