```python
def build_decay(fitting_window, srate, vals=None, x=None):
    if vals is not None:
        coda_height = vals['coda_height']
        peak_decay = np.exp(vals['peak_decay'])
        coda_decay = np.exp(vals['coda_decay'])
    if x is not None:
        coda_height = x[0]
        peak_decay = np.exp(x[1])
        coda_decay = np.exp(x[2])
        
    n = len(fitting_window)
    tlen = n / srate
    t = np.linspace(0, tlen, n)
    logt = np.log(t+1)

    pred_log_signal = coda_height - peak_decay * logt - t * coda_decay
    return pred_log_signal
```
