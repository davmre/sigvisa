```python

%pylab inline
from sigvisa.models.distributions import PiecewiseLinear
import scipy.stats
```

```python
xs_fine = np.linspace(-5, 5, 100)
xs_coarse = np.array([-4, -3, -2, -1.5, -1, -.5, 0, 1.0, 2.0, 2.5, 3.0, 4.0])

lps = scipy.stats.norm.logpdf(xs_fine)
lps_coarse = scipy.stats.norm.logpdf(xs_coarse)-2
```

```python
p = PiecewiseLinear(xs=xs_coarse, lps=lps_coarse)
```

```python
plot(xs_coarse, p.ps)
plot(xs_fine, np.exp(lps))
```

```python
import seaborn as sns

samples1 = np.array([p.sample() for i in range(1000)])
samples2 = scipy.stats.norm().rvs(1000)
```
