```python
from sigvisa import Sigvisa

s = Sigvisa()
cursor = s.dbconn.cursor()
sql_query = "select lebo.lon, lebo.lat, lebo.depth, s.lon, s.lat, leba.phase from leb_origin lebo, leb_assoc leba, static_siteid s where lebo.orid=leba.orid and leba.sta=s.sta"
cursor.execute(sql_query)
arrivals = cursor.fetchall()
```

```python
from collections import defaultdict
from sigvisa.utils.geog import dist_km
distancelists = defaultdict(list)
depthlists = defaultdict(list)

for (elon, elat, edepth, slon, slat, phase) in arrivals:
    distance = dist_km((elon, elat), (slon, slat))
    distancelists[phase].append(distance)
    depthlists[phase].append(edepth)
```
