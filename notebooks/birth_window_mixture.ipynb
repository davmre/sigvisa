```python
def splitpoint_lik(y, x1, x2, idx):
    
    y1 = y[:idx]
    y2 = y[idx:]
    
    
    clf = BayesianRidgeWithCov(compute_score=True, fit_intercept=False, normalize=True)
    clf.fit(x1[:idx, :], y1)
    score1 = clf.scores_[-1]
    clf.fit(x2[idx:, :], y2)
    score2 = clf.scores_[-1]
    return score1+score2

#sl = [splitpoint_lik(y, i) for i in range(5, 45)]
```

```python
def get_fit_window_dist_bayes(wn, peak_time, signal_diff_pos, incr_s=2.0, max_s=60.0, min_s=5.0):

    sidx = int((peak_time-wn.st)*wn.srate)
    jump_idx = int(incr_s * wn.srate)
    min_idx = int(min_s * wn.srate)
    max_idx = min(len(signal_diff_pos) - sidx, int(max_s * wn.srate))

    if sidx >= len(signal_diff_pos)-min_idx:
        return None

    #s = smooth(signal_diff_pos, 15.0)
    s = signal_diff_pos
    lw = np.log(s)[sidx:sidx+max_idx]

    tlen = max_idx/wn.srate
    n = max_idx
    t = np.linspace(0, tlen, n).reshape((-1, 1))
    logt = np.log(t+1.0)
    b = np.ones(t.shape)
    x1 = np.hstack([b, logt, t])
    x2 = np.hstack([b,])

    idx_range = range(min_idx, max_idx-3, jump_idx)
    split_fits = [splitpoint_lik(lw, x1, x2, i) for i in idx_range]
    split_ll = [ll for (ll, clf) in split_fits]
    best_split = idx_range[np.argmax(split_ll)]
    fit_window = signal_diff_pos[sidx:sidx+best_split].copy()
    return split_fits, sidx
```

```python
def build_decay(vals=None, x=None, n=100, srate=1.0):
    if vals is not None:
        coda_height = vals['coda_height']
        peak_decay = -np.exp(vals['peak_decay'])
        coda_decay = -np.exp(vals['coda_decay'])
    else:
        coda_height, peak_decay, coda_decay=x
        peak_decay = -np.exp(peak_decay)
        coda_decay = -np.exp(coda_decay)
        
    tlen = n / srate
    t = np.linspace(0, tlen, n)
    logt = np.log(t+1)

    pred_log_signal = coda_height + peak_decay * logt + t * coda_decay
    return pred_log_signal

ax = do_plot(wn)
pls = build_decay(x=x, n=300)
ax.plot(np.arange(peak_time, peak_time+len(pls)), wn.nm.c+ np.exp(pls))

```
