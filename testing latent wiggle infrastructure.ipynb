```python
from sigvisa.signals.common import Waveform
from sigvisa.plotting.plot import subplot_waveform

def log_plot_wiggle(latent, ax):
    w1 = latent.get_wave().filter('log')
    val = latent._predict_value()
    val[0] = 1
    logval = np.log(val)
    w2 = Waveform(data=logval, stime=latent._tmpl_params['arrival_time'], sta=latent.sta, srate=latent.srate, chan=latent.chan, filter_str='env;' + latent.band)

    subplot_waveform(w1, ax, plot_dets=False, c='black')
    subplot_waveform(w2, ax, plot_dets=False, c='black')
    
    
def debug_plot_all(latent, wn, msg=None):
    if msg is None:
        msg = "round %d" % i
    figure()
    wn.debugging_plot(ax=gca(), plot_mode="full")
    title('%s: observed signal, overall lp %f' % (msg,sg.current_log_p()))
    figure()
    latent.debugging_plot(ax=gca(), plot_mode="predict")
    title('%s: latent signal with predicted shape' % (msg,))
    figure()
    latent.debugging_plot(ax=gca(), plot_mode="wiggle")
    title('%s: wiggle lp %f' % (msg,latent.log_p()))
    figure()
    wn.debugging_plot(ax=gca(), plot_mode="noise")
    title('%s: noise lp %f' % (msg,wn.log_p()))
    figure()
    log_plot_wiggle(latent, gca())
    title('%s: logplot of latent_signal vs predicted shape' % msg)
```

```python
def wiggle_param_step(n_latent):
    noise_std = n_latent.arwm.em.std
    params = n_latent.arwm.params
    
    proposed_std = float(np.exp(np.log(noise_std) + np.random.normal(0, 0.1, 1)))
    #proposed_params = params + np.random.normal(0, 0.1, len(params))

    
    lp_old = n_latent.log_p()

    old_wm = latent.arwm
    
    em = ErrorModel(mean=0.0, std=proposed_std)
    latent.arwm = ARModel(params=params, c=1.0, em = em, sf=n_latent.srate)
    
    lp_new = n_latent.log_p()
    
    u = np.random.uniform()
    if lp_new - lp_old > np.log(u):
        print "updated wiggle params:", proposed_std, params, lp_new-lp_old
        return True
    else:
        latent.arwm = old_wm
        return False
    
def noise_param_step(wn):
    noise_mean = wn.nm.c
    noise_std = wn.nm.em.std
    params = wn.nm.params
    
    proposed_mean = float(np.exp(np.log(noise_mean) + np.random.normal(0, 0.1, 1)))
    proposed_std = float(np.exp(np.log(noise_std) + np.random.normal(0, 0.1, 1)))
    #proposed_params = params + np.random.normal(0, 0.1, len(params))
    
    lp_old = wn.log_p()

    old_nm = wn.nm
    
    em = ErrorModel(mean=0.0, std=proposed_std)
    wn.nm = ARModel(params=params, c=proposed_mean, em = em, sf=wn.srate)
    
    lp_new = wn.log_p()
    
    u = np.random.uniform()
    if lp_new - lp_old > np.log(u):
        print "updated noise params:", proposed_mean, proposed_std, params, lp_new-lp_old
        print "   previous:", noise_mean, noise_std, params
        return True
    else:
        wn.nm = old_nm
        return False
```
