```python
def do_plot(wn, ax=None):
    if ax is None:
      f = plt.figure(figsize=(15,5))
      ax=f.add_subplot(111)
    shape_colors = plot_with_fit_shapes(fname=None, wn=wn, axes=ax, plot_wave=True)
    atimes = dict([("%d_%s" % (eid, phase), wn.get_template_params_for_arrival(eid=eid, phase=phase)[0]['arrival_time']) for (eid, phase) in wn.arrivals()])
    colors = dict([("%d_%s" % (eid, phase), shape_colors[eid]) for (eid, phase) in wn.arrivals()])
    plot_pred_atimes(dict(atimes), wn.get_wave(), axes=ax, color=colors, alpha=1.0, bottom_rel=-0.1, top_rel=0.0)
    return ax

```

```python
from sigvisa.experiments.one_station_templates_test import sample_template

def birth_proposal_signal_prob(birth_helper, nm_type='l1', visualize=False, proposals_per_seed=10):
    lps = [] 
    for seed in range(50):
        wave, templates = sample_template(seed=seed, hardcoded=False, nm_type=nm_type, n_templates=1, srate=1.0)
        tvals = dict([(k, n.get_value()) for (k, n) in templates[0].items()])

        sg = SigvisaGraph(template_model_type="dummy", template_shape="lin_polyexp",
                              wiggle_model_type="dummy", wiggle_family="dummy",
                              phases="leb", nm_type = nm_type, wiggle_len_s = 60.0)
        wn = sg.add_wave(wave)

        tg = sg.template_generator('UA')
        tg.hack_force_mean = np.log(wn.nm.c * 10)

        lp1 = wn.log_p()
        
        if visualize:
            f = plt.figure(figsize=(15, 5))
            ax = f.add_subplot(111)
        
        np.random.seed(seed)
        for i in range(proposals_per_seed):
            lp_old, lp_new, log_qforward, log_qbackward, accept_move, revert_move = birth_helper(sg, wn, laplace_reparam=True, laplace_proposal=False)
            lp2 = wn.log_p()
            
            if visualize:
                do_plot(wn, ax=ax)
                
            revert_move()
            lps.append([lp2-lp1])
            
        if visualize:
            ax.set_title("seed %d" % seed)
            
        print "seed", seed, "probs", lp_new, lp_old, log_qforward, log_qbackward, lp2, lp1
        
    return np.array(lps)

lp1 = birth_proposal_signal_prob(optimizing_birth_helper, proposals_per_seed=1, visualize=True, nm_type='l1')

#lp1 = birth_proposal_signal_prob(optimizing_birth_helper, proposals_per_seed=1, visualize=True, nm_type='ar')

print np.sum(lp1)
```
