```python
from sigvisa import Sigvisa
s = Sigvisa()

from sigvisa.models.distributions import InvGamma
import scipy.optimize

def fit_invgamma(data):
    def nll(x):
        a,b = x
        invg = InvGamma(a,b)
        return -np.sum([invg.log_p(v) for v in data])
    r = scipy.optimize.fmin(nll, x0=[1.0, .5], disp=True)
    ig = InvGamma(*r)
    return ig

cursor = s.dbconn.cursor()
sql_query = "select site, hyperparams from sigvisa_param_model where model_type='param_sin1'"
cursor.execute(sql_query)
sin1_hparams = cursor.fetchall()

sin1_dict = dict([(k, eval(v, {'array': np.array})) for (k,v) in sin1_hparams])
```

```python
noise_vars = [v['noise_var'] for v in sin1_dict.values()]
noise_stds = np.sqrt(noise_vars)
import seaborn as sns
sns.distplot(noise_vars)

ig = fit_invgamma(noise_vars)
x = np.linspace(0, 20, 1000)
pdf = np.exp([ig.log_p(xx) for xx in x])
plot(x, pdf)
```
