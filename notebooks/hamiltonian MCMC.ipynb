```python
def runMH(x0, std=1.0, moves=1000):
    tries = 0
    accepts = 0
    
    x = x0
    samples = []
    lp = rv.logpdf(x)
    
    for i in range(moves):
        tries += 1
        x_new = x + np.random.randn(2)*std
        
        new_lp = rv.logpdf(x_new)
        
        u = np.log(np.random.rand())
        if (new_lp - lp) > u:
            accepts += 1
            x = x_new
            lp = new_lp
            
        samples.append(x)
    print "acceptance rate", float(accepts)/tries
    return np.array(samples)

MHsamples = runMH(np.zeros(2))
```

```python
from matplotlib import animation

from IPython.display import HTML

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))

from tempfile import NamedTemporaryFile

VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=20, extra_args=['-vcodec', 'libx264', '-pix_fmt', 'yuv420p'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)


def animate_samples(samples, time_s=2, decay=0.01):
    # First set up the figure, the axis, and the plot element we want to animate
        
    frames = int(time_s * (1000.0/50))
    samples_per_frame = int(float(len(samples))/frames)
    
    print frames, samples_per_frame

    fig, ax = plt.subplots()
    n = samples.shape[0]
    alphas = np.arange(n)/float(n)
    scat = ax.scatter(samples[:, 0], samples[:, 1], linewidth=0, c=alphas, cmap=plt.get_cmap("gray"))
    
    def setup_plot():
        """Initial drawing of the scatter plot."""
        
        ax.axis([-10, 10, -10, 10])

        # For FuncAnimation's sake, we need to return the artist we'll be using
        # Note that it expects a sequence of artists, thus the trailing comma.
        return scat,
    
    # animation function.  This is called sequentially
    def animate(i):
        
        i_sample = i * samples_per_frame
        i_sample_prev = max(0, (i-1) * samples_per_frame)
        
        alphas = 1 - np.exp( decay* (np.arange(i_sample) - i_sample) )
        #alphas = alphas.reshape((-1, 1))
        #colors = numpy.hstack((numpy.zeros_like(alphas), numpy.zeros_like(alphas), numpy.ones_like(alphas), alphas))
        
        print i_sample
        
        if i_sample > 0:
            scat.set_offsets(samples[:i_sample, :])
            scat.set_array(alphas)

        return scat,
        
    # call the animator.  blit=True means only re-draw the parts that have changed.
    anim = animation.FuncAnimation(fig, animate, blit=True, init_func=setup_plot,
                                   frames=frames, interval=50)

    # call our new function to display the animation
    return display_animation(anim)
    
animate_samples(MHsamples)
```

```python
import numdifftools as nd
J = nd.Jacobian(rv.logpdf)
pdfgrad = lambda z : J(z.flatten()).flatten()

def hmc_step(x, L, eps, force_p=None):
    
    if force_p is not None:
        p = force_p
    else:
        p = np.random.randn(2) # momentum
    current_x = x
    current_p = p
    
    print x, p
        
    # half step for momentum
    p = p + eps * pdfgrad(x)/2
    
    # alternate full steps for position and momentum
    for i in range(L):
        
        if np.isnan(p).any():
            print "warning: rejecting move due to NaN momentum"
            return current_x
            
        # full step for position
        x = x + eps * p
        
        # full step for momentum except on the last iteration
        try:
            p_step = eps * pdfgrad(x)
            p = p + p_step/2
            print i, x, p
            p = p + p_step/2
        except:
            import pdb; pdb.set_trace()
                
        
                
    p = p - p_step/2
    
    
    
    # negate momentum to make the proposal
    p = -p
    
    current_U = -rv.logpdf(current_x)
    current_K = np.sum(current_p**2)/2
    
    proposed_U = -rv.logpdf(x)
    proposed_K = np.sum(p**2)/2
    
    accept_lp = current_U-proposed_U+current_K-proposed_K
    
    
    return x, current_p, p, accept_lp
    
def run_hmc(L, eps, steps=300):
    x = np.zeros(2)
    samples = []
    for i in range(steps):
        
        new_x, oldp, p, accept_lp = hmc_step(x, L, eps)
        if np.random.rand() < np.exp(accept_lp):
            x = new_x
            
        samples.append(x)
    return np.array(samples)

#hmc_samples = run_hmc(5, 0.3)
```
