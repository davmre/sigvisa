```python
def k(x1,x2):
    r = np.linalg.norm(x1-x2, 2)
    cov = (1-r)**3 * (3*r + 1)
    cov = max(0.0, cov)
    return cov

def cov_matrix(X, k):
    return np.array([[k(x1, x2) for x2 in X] for x1 in X])

xs = np.reshape(np.linspace(-25, 25, 101), (-1, 1))
K = cov_matrix(xs, k)
imshow(K)
print K

```

```python
Kinv = inv(K)
imshow(np.abs(Kinv) > .1)
print Kinv
```

```python
evals, evecs = np.linalg.eig(K)
evals = sorted(evals, reverse=True)
print evals[0]/evals[-1]

plot(evals)
```
