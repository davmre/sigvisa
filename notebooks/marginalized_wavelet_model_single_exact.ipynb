```python
from sigvisa.explore.wavelet_matching import unflatten_wavelet_coefs, flatten_wavelet_coefs
from sigvisa.models.noise.armodel.model import ARModel, ErrorModel
from sigvisa.models.templates.lin_polyexp import LinPolyExpTemplateGenerator
import scipy.stats


import pywt
```

```python
def dwt(x, len_s, wtype='db9'):
    cc = unflatten_wavelet_coefs(x, wtype=wtype, len_s=len_s)
    return pywt.waverec(cc, wtype, 'zpd')
    
def idwt(s, wtype='db9', levels=999):
    return flatten_wavelet_coefs(pywt.wavedec(s, wtype, 'zpd'), levels=levels)
```

```python
def innovation_cov(P, len_s, wtype):
    # P is a wxw matrix giving the covariance on the wavelet params
    # want to return HPH' where H is the dwt matrix
    # = H(HP)'
    
    k = P.shape[0]
    
    HP = np.empty((len_s, k))
    
    # enumerating P.T gives rows of P.T, which are cols of P
    for i, col in enumerate(P.T):
        HP[:, i] = dwt(col, len_s, wtype)
    
    # now we want cols of (HP)'
    # which are rows of HP
    HPHt = np.empty((len_s, len_s))
    for i, row in enumerate(HP):
        HPHt[:, i] = dwt(row, len_s, wtype)
    
    return HPHt

```

```python
def signal_cov(w_cov, w_len_idx, ar_mult, ar_noise, template):
    n = len(template)
    Sx = ar_mult.cov_matrix(n)
    Sz = ar_noise.cov_matrix(n)
    
    W = Sx.copy()
    
    AEAt = innovation_cov(w_cov, w_len_idx, 'db9')
    W[:w_len_idx, :w_len_idx] += AEAt
    
    for i in range(n):
        W[:,i] *= template
        W[i,:] *= template
        
    return W + Sz

def signal_mean(w_mean, w_len_idx, template):
    pred_rw = dwt(w_mean, w_len_idx, 'db9')
    pred_wiggle = np.ones((len(template),))
    pred_wiggle[:len(pred_rw)] += pred_rw
    
    pred_signal = template * pred_wiggle
    
    return pred_signal

```

```python
from sigvisa.models.statespace.ar import ARSSM
from sigvisa.models.statespace.compact_support import CompactSupportSSM
from sigvisa.models.statespace.combined_fixed import CombinedSSM
from sigvisa.models.statespace.transient import TransientCombinedSSM
```

```python
N = 256
cc = pywt.wavedec(np.zeros(N,), 'db2', 'zpd')
nb = np.sum([len(l) for l in cc])
bases = np.empty((nb, N))

k = 0
for i,l in enumerate(cc):
    for j in range(len(l)):
        cc[i][j] = 1.0
        bases[k+j, :] = pywt.waverec(cc, 'db2', 'zpd')
        cc[i][j] = 0.0
    k += len(l)

print [len(l) for l in cc]
small_bases = bases[:74, :]
wssm = CompactSupportSSM(small_bases)
```

```python
params = np.array([0.6,0.1 ])
evar = 0.5
arssm = ARSSM(params, evar)

ar_model = ARModel(params, ErrorModel(0, np.sqrt(evar)))

zz = arssm.prior_sample(N)
print arssm.run_filter(zz)

arcov = ar_model.cov_matrix(N)
print scipy.stats.multivariate_normal.logpdf(zz, np.zeros((N,)), arcov)
```

```python
def onescale(k):
    return 1

ssm_wave_ar = CombinedSSM([wssm, arssm], [onescale, onescale])
z = ssm_wave_ar.prior_sample(N)
plot(z)
print ssm_wave_ar.run_filter(z)
```

```python
template = 20*np.sin(np.linspace(-20, 20, N))
components = [(wssm, 15, N, template), (arssm, 0, N+100, None)]
tssm = TransientCombinedSSM(components)

```
