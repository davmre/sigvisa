```python
X = loadtxt('X.txt')
p = np.random.permutation(X.shape[0])
smallX = X[p[:1000]]
smallX = sorted(smallX, key = lambda z : z[0])
import sigvisa.utils.geog as geog
D = np.array([[geog.dist_km((x[0], x[1]), (y[0], y[1])) for x in smallX] for y in smallX])
```

```python
K1 = exp(-.5* D**2)
K10 = exp(-.5* D**2)
K100 = exp(-.5 * D**2/100**2)
eigval1, eigvec1 = eig(K1)
eigval10, eigvec10 = eig(K10)
eigval100, eigvec100 = eig(K100)
```

```python
imshow(K100)
figure()
plot(real(eigval100))
```

```python

ev = eigval100
ev[100:] = 0
lrank = np.dot(eigvec100, np.dot(np.diag(ev), eigvec100.T))
imshow(lrank)
```

```python
print K100[800, 200]
print lrank[800, 200]
imshow(K100-lrank)
figure()
imshow(log(K100))
figure()
imshow(log(lrank))
```
